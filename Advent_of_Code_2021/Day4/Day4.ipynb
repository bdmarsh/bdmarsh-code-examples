```python
import pandas as pd
import numpy as np
```

```python
with open('Day4_Input.txt') as f:
    input = f.read()
    f.close
```

```python
input = input.split(',')
input = list(map(int, input))
```

```python
#get the boards into an array of subarrays

#read in the text as string to keep the leading zeros
lines = pd.read_csv('Day4_Boards.txt', header=None, sep='\n\n', dtype= 'str', engine='python')
lines = lines[0].str.split(' +', expand=True) #split on one or more spaces
                                              #(some numbers were a single digit with a leading space)

#convert to an array of integers
arr = np.asarray(lines, dtype='int32')
arr = np.split(arr, 100) #split into subarrays
arr
```

```python
win = False
i=0

while win == False:
    entry = input[i]
    
    for j in range(100):
        arr[j] = np.where(arr[j]==entry, 0, arr[j])
        row_sum = arr[j].sum(axis=1)
        col_sum = arr[j].sum(axis=0)
        total = arr[j].sum() * entry
        if np.any(row_sum == 0):
            win = True
            break

        if np.any(col_sum == 0):
            win = True
            break
    i += 1
print(total)
```
